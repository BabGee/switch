## LOGIN GOOGLE

### Section 1: Use case

- User clicks on LOGIN WITH GOOGLE
- User selects Google account.
- If user profile is found login user then redirect to DASHBOARD.
- if user profile not found; create user and redirect to select industry page 
- User selects industry
- User is logged in and redirected to dashboard.

### Section 3: pseudocode

 LOGIN SOCIAL
    
- get_social_profile(returns payload with user profile details(email, first_name etc)) 
- get_update_profile(returns user profile if found(generates trigger-profile_found) else creates user profile(generates trigger-email_not_registered))
- get_section(trigger- email_not_registred) - Select Industry page
- login (trigger - profile_found)
- redirect (triggers- active_password, profile_found)

N/B: Assuming the user profile is found so the service commands that will execute(trigger- profile_found) are:
- get_social_profile(works)
- get_update_profile(works)
- login(fails - returns empty response) generates payload expired_password
- redirect(triggers- active_password, profile_found) - won't run
